```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Load dataset
df = pd.read_csv('HousING.csv') 
print(df.head())
print(df.info())
df = df.dropna()

```

```python
X_simple = df[['area']]     
y_simple = df['price']      
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_simple, y_simple, test_size=0.2, random_state=1)

model_s = LinearRegression()
model_s.fit(X_train_s, y_train_s)

y_pred_s = model_s.predict(X_test_s)

print("\nSimple Regression Results:")
print("MAE:", mean_absolute_error(y_test_s, y_pred_s))
print("MSE:", mean_squared_error(y_test_s, y_pred_s))
print("R2:", r2_score(y_test_s, y_pred_s))
```

```python

plt.scatter(X_test_s, y_test_s, color='blue', label='Actual')
plt.plot(X_test_s, y_pred_s, color='red', linewidth=2, label='Predicted')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

```python
X_multi = df[['area', 'bedrooms', 'bathrooms']]
y_multi = df['price']

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multi, y_multi, test_size=0.2, random_state=1)

model_m = LinearRegression()
model_m.fit(X_train_m, y_train_m)

y_pred_m = model_m.predict(X_test_m)

print("\nMultiple Regression Results:")
print("MAE:", mean_absolute_error(y_test_m, y_pred_m))
print("MSE:", mean_squared_error(y_test_m, y_pred_m))
print("R2:", r2_score(y_test_m, y_pred_m))

print("Intercept:", model_m.intercept_)
print("Coefficients:", model_m.coef_)
```
